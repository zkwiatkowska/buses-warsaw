# Profiling

This is an example how to run profiler for the functions in the library if you use jupyter notebooks with your code. You can use API key provided for debugging this project, but in the long run you should set up your own API key on the UM Warsaw website [here](https://api.um.warszawa.pl/). 

```python
import cProfile
from bwaw.api.requests import get_bus_stops_coordinates
```

```python
API_KEY = '5fbe79ed-1f5b-4019-ab03-641443842d8b'
```

Profiling in terms of total execution time:

```python
cProfile.run('get_bus_stops_coordinates(api_key=API_KEY)', sort='tottime') 
```

Profiling in terms of number of calls:

```python
cProfile.run('get_bus_stops_coordinates(api_key=API_KEY)', sort='ncalls')
```
